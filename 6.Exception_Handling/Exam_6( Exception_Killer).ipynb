```python
# Exception Killer #
#==================#

class Calculate():  #base class
    def __init__(self):     #constructor with data members
        self.acno=None
        self.name=None
        self.money=None

        self.d_money=None
        self.w_money=None

    def Create_account(self):   #method for create account
        
        self.acno=int(input("Enter your account no:"))
        self.name=input("Enter your name:")
        self.money=0

        return self.acno,self.name      #return account number and name
    

    def Display_data(self):     #method for display account
        print(f"A/C no. :{self.acno} \nName :{self.name} \nMoney :{self.money}")


    def deposite(self):     #method for deposite money
        self.d_money=int(input("Enter deposite money:"))

        self.money+=self.d_money    #entered amount plus in default amount
        
        print(f"{self.d_money} Rs. Deposited Successfully")
        print(f"New balence :{self.money}")     #show current amount


    def withdraw(self):     #method for withdraw money
        self.w_money=int(input("Enter withdrow money:"))

        try:    #Exception handling for big amount from current amount
            if self.w_money>self.money:
               raise ValueError("Your amount is big fom your curreunt") 
            
        except ValueError as e: #show current amount with error meaasge
            print(f"{e} balence :{self.money}")

        else:    
            self.money-=self.w_money    #entered amount minuse in default amount
            
            print(f"{self.w_money} Rs. Withdrow Successfully")
            print(f"New balence :{self.money}")#show current amount

            
cal=Calculate()     #object of class

while True:

    print("\n*** Please Select An Operation ***\n")
    print("1. Create Account")
    print("2. Exit")

    choise=int(input("Enter your choise:"))
    match choise:

        case 1:     #create account
            cal.Create_account()
            print(f"Account Created Successfully....")
                
            while True:
        
                print("\n*** OTHER FACELITIES ***\n")
                print("1. Deposite Funds")
                print("2. Withdrow Funds")
                print("3. Check Balence")
                print("4. Exit\n")
                inner_choise=input("Enter your choise:")

                try:        #Exception handling for only integer choise
                    inner_choise=int(inner_choise)

                    match inner_choise:
                        case 1:
                            cal.deposite()  #deposite money

                        case 2:
                            cal.withdraw()  #withdraw money

                        case 3:
                            cal.Display_data()  #display account
                    
                        case 4:
                            print("Exiting From Fecilities.....")
                            break

                        case _:
                            print("Wrong choise!!!")
                    
                except Exception as e:  #give error for other entered value
                    print(e,"Error: To Perform Operation To Give Only Integers...")


        case 2:     #Exit programm
            print("Thank you .... Exiting Programm...")
            break

        case _:
            print("Wrong choise!!!")
```
