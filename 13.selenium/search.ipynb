```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

search="python tutorial"

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.get(f"https://www.wikipedia.org/")
time.sleep(2)


elem = driver.find_element(By.ID,"searchInput")
time.sleep(2)

elem = driver.find_element(By.ID,"searchInput")
elem.send_keys(search)
time.sleep(2)

elem = driver.find_element(By.ID,"searchInput")
elem.send_keys(Keys.RETURN)
time.sleep(2)



import requests
from bs4 import BeautifulSoup
import pandas as pd

Title=[]
Link=[]

url="https://en.wikipedia.org/wiki/Special:Search?go=Go&search=python+tutorial&ns0=1"

request=requests.get(url)
# print(request.status_code)

soup=BeautifulSoup(request.text,"html.parser")



title=soup.find_all("div",class_="mw-search-result-heading")

for i in title:
    tt=i.text
    Title.append(tt)

for i in Title:
    print(i)


links=soup.find_all("div",class_="mw-search-result-heading")

for i in links:
    link=i.a['href'] 
    final=url+link
    Link.append(final)

for i in Link:
    print(i)

print(len(Title))
print(len(Link))


driver.quit()
  
```
