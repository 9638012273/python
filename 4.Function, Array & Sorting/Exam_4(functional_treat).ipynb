```python
# FUNCTIONAL TREAT #
#==================#

print("WELCOME TO THE DATA ANALYZER AND TRANSFORMER PROGRAMM !\n")
arr=[]      #empty arry to store values

while True:     #loop continue while programm not break

    print("**** MAIN MENU ****\n")

    print("1. Input Data")
    print("2. Display Data Summary (Built-in function)")
    print("3. Calculate Fectorial (Recursion)")  
    print("4. Calculate Fibonacci series (Recursion)")
    print("5. Filter Data By Threshold (lambda function)")      
    print("6. Sort data")
    print("7. Display Dataset Statistics (Return Multiple Values)")
    print("8. Other Opstions (Find duplicates , Unique values)")
    print("9. Exit Programm\n")
    choise=int(input("\n Enter your choise:"))      

    match choise:       

        case 1:       #CASE 1 TAKE ARRAY FROM USER

            print("=== Add Data ===\n")
            print("1. 1D ARRAY")
            print("2. 2D ARRAY")
            arr_choise=int(input("Enter array choise:"))

            if arr_choise==1:

                data=input("Enter data for a 1D array (Seprate By Space):")
                arr=[int(item) for item in data.split( )]   
                print("DATA INSERT SUCCESSFULLY....\n\n")   #first string convert in int and using 
                                                            #split function take elements from data using 
                                                            # for loop and use to space
            elif arr_choise==2:

                rows=int(input("Enter rows for insert value in 2D array:"))

                for i in range(1,rows+1):       #loop continue for range of rows
                    data=input(f"Enter data for row{i} (Seprate By Space):")
                    arr.append([int(item) for item in data.split( )])
                print("DATA INSERT SUCCESSFULLY....\n\n")


        case 2:     #CASE 2 FOR DISPLAY ALL DATA INFORMATION
            if len(arr)==0:   
                print("ARRAY IS EMPTY ... FIRST ENTER DATA...")
            
            else:           
                print(" === DATA SUMMMARY === \n")
                if isinstance(arr[0],list):  # check array is 2d or 1d

                    print("2D ARRAY:")
                    print(arr,"\n")

                    length=0
                    minimum=arr[0][0]   #dive first element value
                    maximum=0
                    total_sum=0

                    for row in arr:
                        length+=len(row)        #find length
                        for value in row:
                            if value<minimum:   #find minimum
                                minimum=value
                            if value>maximum:   #find maximum
                                maximum=value

                            total_sum+=value    #find sum
                            

                    print("- Total elements:",length)
                    print("- Minimum value:",minimum)
                    print("- Maximum value:",maximum)
                    print("- Sum of all values:",total_sum)
                    print(f"- Average value:{(total_sum/length):.2f}\n")    #average 


                else:       #1D array
                  
                    print("- 1D ARRAY :",arr,"\n")
                    print("- Total elements:",len(arr))
                    print("- Minimum value:",min(arr))
                    print("- Maximum value:",max(arr))
                    print("- Sum of all values:",sum(arr))

                    def cal_avg(arr):       #find average using udf function
                        sum=0
                        cnt=0
                        for i in arr:
                            cnt+=1
                            sum+=i

                        return (sum/cnt)

                    average=cal_avg(arr)
                    print(f"- Average value:{average:.2f}\n")


        case 3:     #CASE 3 FOR FIND FECTORIAL  
         
                print(" === FIND FECTORIAL === \n")

                # FIND FECTORIAL USING RECURSIVE FUNCTION

                def calculate(num):
                    if num<=1:           # CONDITION NUM LESSTHEN = 1 
                        return 1
                    else:
                        return num*calculate(num-1)  #FUNCTION CALL IT SELF AND CALCULATE FECTORIAL
                    

                num=int(input("Enter number to calculate its fectorial"))

                fectorial=calculate(num)    # CALL FUNCTION
                print(f"Fectorial on {num} is :{fectorial}\n")


        case 4:     #CASE 4 FOR FIND FIBONACCI SERIES
       
                print(" === FIND FIBONACCI SERIES === \n")

                # recirsive fibonacci series

                def fab(n):
                    if n<=0:
                        print("Enter positive number...")
                    elif n==1:
                        return 0
                    elif n==2:
                        return 1
                    else:
                        return fab(n-1)+fab(n-2)    #ex. num=5 fab(4)+fab(3)=8

                num=int(input("Enter number for calculate febonacci series:"))
                fab_list=[]

                for i in range(1, num+1):   #loop continue for range
                    fab_list.append(fab(i))  #call function and value append in list

                print(fab_list) #print list
                print("\n")
           

        case 5:       #CASE 5 FILTER DATA FROM THRESHOLD VALUE
            if len(arr)==0:   
                print("ARRAY IS EMPTY ... FIRST ENTER DATA...")
            
            else:  
                th_value_list=[]    #empty list

                print(" === FILTER DATA BY THRESHOLD  === \n")     
                if isinstance(arr[0],list): #check array is 1D or 2D

                    th_value=int(input("Enter threshold value to filter data:"))    #take threshold value rom user

                    threshold=lambda arr,th_value:[value for row in arr for value in row if value>th_value]
                    #lambada function for threshold value
                    th_value_list=threshold(arr,th_value)   #call lambda

                    print(f"Filtered Data (value >= {th_value}):")
                    print(th_value_list,"\n")

                else: 

                    th_value=int(input("Enter threshold value to filter data:"))

                    threshold=lambda arr,th_value:[value for value in arr if value>th_value]
                    #lambada function for threshold value
                    th_value_list=threshold(arr,th_value)    #call lambda

                    print(f"Filtered Data (value >= {th_value}):")
                    print(th_value_list,"\n")


        case 6:       #CASE 6 SORTIND ARRAY VALUES
            if len(arr)==0:   
                print("ARRAY IS EMPTY ... FIRST ENTER DATA...","\n")
            
            else:           
                print(" === SORTING DATA  === \n")
                print("1. Ascending")
                print("2. Descending")

                sorting_choise=int(input("Enter sorting choise:"))

                if isinstance(arr[0],list):

                    if sorting_choise==1:       #Ascending sorted 2d array data
                        a_sort=sorted([value for row in arr for value in row])
                        print("Ascending:",a_sort,"\n")


                    elif(sorting_choise==2):       #Descending sorted 2d array data
                        a_sort=sorted([value for row in arr for value in row],reverse=True)
                        print("descending:",a_sort,"\n")

                    else:
                        print("Wrong choise!!!")

                else:

                    if sorting_choise==1:       #Ascending sorted 1d array data
                        arr.sort()
                        print("Ascending:",arr,"\n")

                    elif(sorting_choise==2):       #Descending sorted 1d array data
                        arr.sort(reverse=True)
                        print("descending:",arr,"\n")

                    else:
                        print("Wrong choise!!!")

        case 7:       #CASE 7 FOR DATASET STETISTICS
            if len(arr)==0:   
                print("ARRAY IS EMPTY ... FIRST ENTER DATA...")
            
            else:           
                print(" === DATASET STETISTICS === \n")

                if isinstance(arr[0],list):
                    def calculate(arr):       #udf function usinf *args

                        length=0
                        minimum=arr[0][0]
                        maximum=0
                        total_sum=0

                        for row in arr:
                            length+=len(row)
                            for value in row:
                                if value<minimum:   #find minimum
                                    minimum=value
                                if value>maximum:   #find maximum
                                    maximum=value

                                total_sum+=value    #total sum
                        return length,minimum,maximum,total_sum

                    len_2d,min_2d,max_2d,sum_2d=calculate(arr)

                    print("- Minimum value:",min_2d)
                    print("- Maximum value:",max_2d)
                    print("- Sum of all values:",sum_2d)
                    print(f"- Average value:{(sum_2d/len_2d):.2f}\n")   #average

                else:

                    def cal_avg(arr):   # udf for find average of 1D array
                        cnt=0
                        sum=0
                        for i in arr:
                            cnt+=1
                            sum+=i
                            total_avg=(sum/cnt)
                        return total_avg,sum

                    average,sum_of_arr=cal_avg(arr)

                    print("- Minimum value:",min(arr))  #minimum
                    print("- Maximum value:",max(arr))  #maximum
                    print("- Sum of all values:",sum_of_arr)    #total sum
                    print(f"- Average value:{average:.2f}\n")   #average


        case 8:       #CASE 8 FOR FIND DUPLICATES AND UNIQUE VALUE FROM ARRAY
            if len(arr)==0:   
                print("ARRAY IS EMPTY ... FIRST ENTER DATA...","\n")
            
            else:           
                print(" === Other Functionality === \n")
                print("1. Find Duplicates")
                print("2. Find Unique")

                sorting_choise=int(input("Enter your choise:"))
                if isinstance(arr[0],list):

                    if sorting_choise==1:   #find duplicates value 
                        def cal(arr):
                            dict={}
                            d_list=[]

                            for row in arr:
                                for value in row:
                                    if value in dict:   #check frequency of array values
                                        dict[value]+=1
                                    else:
                                        dict[value]=1

                            for value,count in dict.items():
                                if count>1: #if array value frequency greater then 1
                                    d_list.append(value)
                            return d_list

            
                        duplicate=cal(arr)
                        print("dupicates value:",duplicate)


                    elif(sorting_choise==2):    #find unique value using set
                        unique=list(set(value for rows in arr for value in rows))
                        print("Unique values:",unique)

                    else:
                        print("Wrong choise!!!")

                else:       #for 1D array

                    if sorting_choise==1:
                        def cal(arr):

                            dict={}
                            d_list=[]
                            for value in arr:
                                if value in dict:   #check frequency of array values
                                    dict[value]+=1
                                else:
                                    dict[value]=1

                            for value,count in dict.items():
                                if count>1:  #if array value frequency greater then 1
                                    d_list.append(value)
                            return d_list

                                
                        duplicate=cal(arr)
                        print("dupicates value:",duplicate)

                    elif(sorting_choise==2):    #find unique value using set
                        unique=list(set(arr))
                        print("Unique values:",unique)

                    else:
                        print("Wrong choise!!!")


        case 9:        #CASE 9 FOR EXIT
            print("THANK YOU For Coming...Exiting System...!!!")
            break       #break programm
        

        case _:        #CASE _ FOR INVELID CHOISE

            print("INVELID CHOISE...!!!")

```
