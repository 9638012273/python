```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
from datetime import datetime
import os
from PIL import Image

f_name=input("Enter Your First name:")
l_name=input("Enter Your Last name :")
mobile=input("Enter Your Mobile no.:")
email=input("Enter Your Email:")
street=input("Enter street address:")
city=input("Enter Your city:")
state=input("Enter Your state:")
gender=input("Enter Your gender:")
dob=input("Enter your Birth Date (DD-MM-YYYY):")
visit_date=input("Enter your Hoapital Visit Date (DD-MM-YYYY):")
problem=input("Enter your problem:")

date= datetime.strptime(dob, "%d-%m-%Y")
dob_day = date.day
dob_month = date.month
dob_year = date.year

datevisit= datetime.strptime(visit_date, "%d-%m-%Y")
visit_day = date.day
visit_month = date.month
visit_year = date.year



service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.get(f"https://form.jotform.com/umaretiyamiit/personal-data-for-health-checkup")
time.sleep(1)


elem = driver.find_element(By.ID,"first_5")
elem.send_keys(f_name)
time.sleep(1)

elem = driver.find_element(By.ID,"last_5")
elem.send_keys(l_name)
time.sleep(1)

elem = driver.find_element(By.ID,"input_6_full")
elem.send_keys(mobile)
time.sleep(1)

elem = driver.find_element(By.ID,"input_8")
elem.send_keys(email)
time.sleep(1)

elem = driver.find_element(By.ID,"input_4_addr_line1")
elem.send_keys(street)
time.sleep(1)

elem = driver.find_element(By.ID,"input_4_city")
elem.send_keys(city)
time.sleep(1)

elem = driver.find_element(By.ID,"input_4_state")
elem.send_keys(state)
time.sleep(1)

elem = driver.find_element(By.ID,"input_21")
elem.send_keys(gender)
time.sleep(1)

elem = driver.find_element(By.ID,"input_22_day")
elem.send_keys(dob_day)
time.sleep(1)
elem = driver.find_element(By.ID,"input_22_month")
elem.send_keys(dob_month)
time.sleep(1)
elem = driver.find_element(By.ID,"input_22_year")
elem.send_keys(dob_year)
time.sleep(1)

elem = driver.find_element(By.ID,"day_15")
elem.send_keys(visit_day)
time.sleep(1)
elem = driver.find_element(By.ID,"month_15")
elem.send_keys(visit_month)
time.sleep(1)
elem = driver.find_element(By.ID,"year_15")
elem.send_keys(visit_year)
time.sleep(1)

elem = driver.find_element(By.ID,"input_24")
elem.send_keys(problem)

button=driver.find_element(By.ID,"input_25")
elem= button.find_element(By.XPATH, "//button[@type='submit']")
elem.send_keys(Keys.RETURN)

Path=os.path.join(os.getcwd(),'healthdata_1.png')
driver.save_screenshot(Path)

driver.quit()

print("\n\t\t==================================================================")
print("\t\t\t \t HEALTH CARE DATA RECIVED ...")
print("\t\t==================================================================")
print(f"\t\t\t *** {f_name},{l_name} -->")
print(f"\t\t\t Mobile no. : {mobile}")
print(f"\t\t\t Email  Id. : {email}")
print(f"\t\t\t Street Add.: {street}")
print(f"\t\t\t City       : {city}")
print(f"\t\t\t State      : {state}")
print(f"\t\t\t Gender     : {gender}")
print(f"\t\t\t DOB        : {dob}")
print(f"\t\t\t Visit date : {visit_date}")
print(f"\t\t\t Problem    : {problem}")
print("\t\t==================================================================")
print(f"\t\tForm Fill-up successfully....\n")
  

current_directory = os.getcwd()
image_path = os.path.join(current_directory, 'healthdata_1.png')
img = Image.open(image_path)
img.show()
```
