```python
# Final_Exam_1 #
#==============#
#https://docs.google.com/document/d/1lQiL_Rquds7xirqJNqgnPZtA1VC0ue3-4EKQmFD-C_s/edit?tab=t.0

print("*******************************************************")
print("Welcome to the personal helth tracker and BMI calulator\n")

print("==================================") #bmi classification
print("BMI Classification")
print("==================================")
print("Under weight: BMI < 18.5")
print("Normal weight: 18.5 <= BMI < 24.9")
print("Over weight: 25<= BMI < 29.9")
print("Obese: BMI >= 30")
print("==================================")

mainlist=[]     #empty list
dict={}         #empty dict

test=True

while test:  #loop continut while test is true
    print("\n===  CHOSE OPERATIONS..  ===\n")   #operations

    print("1. Add A New Health Record")
    print("2. View All Health Record")
    print("3. Update A Health Record")
    print("4. Delete A Health Record")
    print("5. Calculate BMI And Get Health Advive")
    print("6. Analyze Health Trends")
    print("7. Exit\n")

    choise=int(input("\nEnter your choise:"))   #take choise from user

    match choise:
    #For add a new record
        case 1:
            print("*** Add A New Health Record ***")
 
            id=input("Enter Id:")
            name=input("Enter Name:")
            date=input("Enter Date (DD/MM/YYYY):")
            weight=float(input("Enter Weight (kg):"))      #get health details from user
            height=float(input("Enter Height (m):"))
            activity=input("Enter Activity Description:")

            bmi_calculation=weight/(height**2)      #calculate bmi
            bmi=float("%.1f"%bmi_calculation)       #.1f for one integer after point

            if bmi<18.5:        #calculate bmi with weight
                print(f"\n BMI:{bmi} (Under weight)")
            elif 18.5<= bmi <24.9:
                print(f"\n BMI:{bmi} (Normal weight)")
            elif 25<= bmi <=29.9:
                print(f"\n BMI:{bmi} (Over weight)")
            else:
                print(f"\n BMI:{bmi} (Obese)")

            dict[id]={   
                "ID":id,        #data convert in dictionary and id is key of dictionary 
                "Name":name,
                "Date":date,
                "Weight":weight,
                "Height":height,
                "Activity":activity,
                "BMI":bmi,
            }
            mainlist.append(dict[id]) # dictionary append in main list(add)

            print("\nDATA INSERT SUCCESSFULLY....\n")

    #for show all record from mainlist
        case 2:
            if len(mainlist)==0:    #main list is empty
                print("MAIN LIST IS EMPTY ... FIRST ENTER HEALTH DATA...")
            
            else:
                print("*** All HEALTH RECORDS \n")

                for i,record in enumerate(mainlist):    #enumerate for count records
                    print(f"* RECORD :- {i+1}")
                    print("--------------")

                    for key,value in record.items():    #when key==bmi skip this step
                        if key=="BMI":
                            bmi=value
                            continue
                        print(f" {key}:{value}")

                    #calculate bmi with weight and print it
                    if bmi<18.5:
                        print(f" BMI:{bmi} (Under weight)")
                    elif 18.5<= bmi <24.9:
                        print(f" BMI:{bmi} (Normal weight)")
                    elif 25<= bmi <=29.9:
                        print(f" BMI:{bmi} (Over weight)")
                    else:
                        print(f" BMI:{bmi} (Obese)")

                    print("\n")

    # for update values using id
        case 3:
            if len(mainlist)==0:
                print("MAIN LIST IS EMPTY ... FIRST ENTER HEALTH DATA...")
            
            else:
                user_id=input("Enter Id to Update Health details :")
                
                for record in mainlist:
                    if record["ID"]==user_id:
                        #id match when programm continue

                        print("\n*** UPDATE DETAILS ***\n")

                        print("1. Chenge ID")
                        print("2. Chenge Name")
                        print("3. Chenge Date")      #opstion for students 
                        print("4. Chenge Weight")
                        print("5. Chenge Height")
                        print("6. Chenge Activity")
                        print("7. Exit")

                        update_choise=int(input("\n Enter your choise:"))

                    #update id
                        if update_choise==1:
                            for record in mainlist:          #match id in dictionary
                                if record["ID"]==user_id:     

                                    update_id=input("Update Health ID:")
                                    record["ID"]=update_id  #update id via update_id
                                    print("SELECTED ID DETAILS UPDATED SUCCESSFULLY....\n")

                                else:
                                    print("id not match in mainlist.....")

                    #update name
                        elif update_choise==2:
                            for record in mainlist:          #match id in list dictionary
                                if record["ID"]==user_id:     

                                    update_name=input("Update Name:")
                                    record["Name"]=update_name     #update name via update_id
                                    print("Selected ID Name details updated successfully....\n")

                                else:
                                    print("id not match in mainlist.....")

                    #update date
                        elif update_choise==3:
                            for record in mainlist:          #match id in list dictionary
                                if record["ID"]==user_id:     

                                    update_date=input("Update Date(DD/MM/YYYY):")
                                    record["Date"]=update_date     
                                    print("Selected ID Date details updated successfully....\n")

                                else:
                                    print("id not match in mainlist.....")

                    #update weight
                        elif update_choise==4:
                            for record in mainlist:          #match id in list dictionary
                                if record["ID"]==user_id:     

                                    update_weight=float(input("Update Weight(kg):"))
                                    record["Weight"]=update_weight 
                                    print("Selected ID Weight details updated successfully....\n")

                                #update bmi record with updated weight
                                    update_height=record["Height"]
                                    update_bmi=update_weight/(update_height**2)
                                    record["BMI"]=float("%.1f" %update_bmi)   
                                    
                                else:
                                    print("id not match in mainlist.....")

                    #update height
                        elif update_choise==5:
                            for record in mainlist:          #match id in list dictionary
                                if record["ID"]==user_id:     

                                    update_height=float(input("Update height(m):"))
                                    record["Height"]=update_height
                                    print("Selected ID Height details updated successfully....\n")

                                #update bmi record with updated height
                                    update_weight=record["Weight"]
                                    update_bmi=update_weight/(update_height**2)
                                    record["BMI"]=float("%.1f" %update_bmi)     
                                    
                                else:
                                    print("id not match in mainlist.....")

                    #update Activity
                        elif update_choise==6:
                            for record in mainlist:          #match id in list dictionary
                                if record["ID"]==user_id:     

                                    update_activity=input("Update Activity:")
                                    record["Activity"]=update_activity    
                                    print("Selected ID Activity details updated successfully....\n")

                                else:
                                    print("id not match in mainlist.....")

                        else:
                            print("WRONG CHOISE...\n")

                    else:
                        print("ID not match in mainlist.....")

    #for delete selected id record
        case 4:
            if len(mainlist)==0:
                print("MAIN LIST IS EMPTY ... FIRST ENTER HEALTH DATA...")
            
            else:
                del_id=input("Enter Id to Delete Health details :")
                
                #del_id match to records
                for record in mainlist:
                    if record["ID"]==del_id:
                        mainlist.remove(record)     #delete matched id record
                        print(f"Selected ID:{del_id} deleted from mainlist....")
                    else:
                        print("ID not match in mainlist.....")
      
    # for calculate BMI with advice 
        case 5:
            if len(mainlist)==0:
                print("MAIN LIST IS EMPTY ... FIRST ENTER HEALTH DATA...")
            
            else:
                print("*** BMI CALCULATOR AND ADVICER \n")
                advice_id=input("Enter id for calculate BMI and get Advice")

                for record in mainlist:          #match id in dictionary
                    if record["ID"]==advice_id:
                        print("--------------------")
                    #get name,weight and height for selected id
                        print(f"Name:{record["Name"]}")
                        print(f"Weight:{record["Weight"]}")
                        print(f"Height:{record["Height"]}")

                    #advice for under weight person 
                        if bmi<18.5:
                            print(f"\n*BMI:{bmi} (Under weigth)")
                            print("* Advice:")
                            print("-------------------------------------------")
                            print("1. Eat More Often to Gain Weight")
                            print("2. Choose Low Volume Foods to Gain Weight")
                            print("3. Get Protein At Every Meal to Gain Weight")
                            print("4. Cook With Healthy Fats to Gain Weight")
                            print("5. Track Your Intake to Gain Weight\n")

                    #advice for normal weight person 
                        elif 18.5<= bmi <24.9:
                            print(f"\n*BMI:{bmi} (Normal weigth)")
                            print("* Advice:")
                            print("-------------------------------------------")
                            print("1. Do your work as do as normaly")
                            print("2. Take simple and sort food")
                            print("3. Try new things")
                            print("4. Go to sports games")
                            print("5. Do some Exercises for normal health\n")

                    #advice for under over person 
                        elif 25<= bmi <=29.9:
                            print(f"*BMI:{bmi} (Over weigth)")
                            print("* Advice:")
                            print("-------------------------------------------")
                            print("1. Do some hard work for fetloss")
                            print("2. Take simple food for strong digestion system")
                            print("3. Improve your Working persionality")
                            print("4. Go to GYM and take some exercise ")
                            print("5. Do things for weightloss\n")

                    #advice for obese person 
                        else:
                            print(f"*BMI:{bmi} (Obese)")
                            print("* Advice:")
                            print("-------------------------------------------")
                            print("1. Don't bother with crash diets")
                            print("2. Look out for hidden sugar")
                            print("3. Get your heart pumping")
                            print("4. Simply move more")
                            print("5. Get the right support\n")
                        
                    else:
                        print("ID not match in mainlist.....")

    # for analyze health trends(summary) 
        case 6:     
            if len(mainlist)==0:
                print("MAIN LIST IS EMPTY ... FIRST ENTER STUDENT DATA...")
            
            else:
                print("*** ANALYZE HEALTH TRENDS ***\n")

            # give summary of all records like...id,name,weight,height and bmi
                for i,record in enumerate(mainlist):
                    print(f"* RECORD :- {i+1}")
                    print("--------------")
                #take records from dictionary
                    print(f" ID:{record["ID"]}")
                    print(f" Name:{record["Name"]}")
                    print(f" Weight:{record["Weight"]}")
                    print(f" Height:{record["Height"]}")

                    if bmi<18.5:
                        print(f" BMI:{bmi} (Under weight)")
                    elif 18.5<= bmi <24.9:
                        print(f" BMI:{bmi} (Normal weight)")
                    elif 25<= bmi <=29.9:
                        print(f" BMI:{bmi} (Over weight)")
                    else:
                        print(f" BMI:{bmi} (Obese)")
                    print("\n")

     #for exit
        case 7:    
            print("Thank You for coming... Exiting Programm...\n")
            break

        case _:
            print("WRONG CHOISE.....!!!\n")

    # yn choice for if user enter yes programm will continue else break and exit
    yn_choice=input("Do you want to continue(yes / no):")
    if yn_choice=="yes":
        test=True
    elif yn_choice=="no":
        print("Thank You for coming... Exiting Programm...")
        test=False
    else:
        print("Enter only yes... or no ...")
        test=False
```
