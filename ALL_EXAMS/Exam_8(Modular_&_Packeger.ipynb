```python
# Moduler And Packger #
#=====================#

from datetime import datetime

print("\n=================================")
print(" Welcome To Multi-Utility Toolkit")
print("=================================")
while True:     #LOOP CONTINUE WHILW TRUE

    print("===============================================")
    print("  ***   Chose an option:")
    print("-----------------------------------------------")
    print("   1. Datetime and Time Operations")
    print("   2. Methemetical Operations")                        #OPSTIONS
    print("   3. Random Date Generation")
    print("   4. Generate Unique Identifiers (UUID)")
    print("   5. File Operations (Custom Module)")
    print("   6. Explore Module Attributes (dir())")
    print("   7. Exit")
    print("===============================================\n")

    main_choise=int(input("Enter your choise:"))

    match main_choise:      #SWITCH CASE FOR MATCH CHOISES
    
    #CASE 1 FOR FILE SYSTEM OPERATIONS
        case 1:
            while True:
                print("\n==============================================")
                print("  ***   Enter Your Choise:")
                print("   1. Display Current Date and Time")
                print("   2. Calculate Diffrent Between Two Date/times")
                print("   3. Format Date Into Custom Formet")
                print("   4. Stopwatch")
                print("   5. Countdown Timer")           
                print("   6. Back To Main Menu\n")

                File_choise=int(input("Enter Your Choise:"))
                match File_choise:
                # show current date and time
                    case 1:
                       
                        print(datetime.now().strftime("Current date: %Y-%m-%d || and Time: %H:%M:%S"))
                        
                #calculate diffrent between two dates 
                    case 2:
                        string1=input("Enter The First Date(YYYY-MM-DD):")
                        string2=input("Enter The second Date(YYYY-MM-DD):")

                        date1=datetime.strptime(string1,"%Y-%m-%d")
                        date2=datetime.strptime(string2,"%Y-%m-%d")

                        Difference=(date2-date1).days
                        print(f"Difference:{Difference} days")


                        
                #Formet date in to custom formet
                    case 3:

                        string=input("Enter date (YYYY-MM-DD)")

                        date=datetime.strptime(string,"%Y-%m-%d")
                        print("Chose An Option:")
                        print("1. MM-DD-YYYY")
                        print("2. DD-MM-YYYY")
                        print("3. YYYY-DD-MM")
                        choise_date=int(input("Enter your choise"))

                        if choise_date==1:
                            print("\n................")
                            print("  *. MM-DD-YYYY")
                            print(datetime.strftime(date,"Date:%m-%d-%Y"))

                        elif choise_date==2:
                            print("\n................")
                            print("  *. DD-MM-YYYY")
                            print(datetime.strftime(date,"Date:%d-%m-%Y"))

                        elif choise_date==3:
                            print("\n................")
                            print("  *. YYYY-DD-MM")
                            print(datetime.strftime(date,"Date:%Y-%d-%m"))

                        else:
                            print("Invelid choise...")
                        

                #stopwatch
                    case 4:
                    
                        while True:
                            print("Enter your choise:")
                            print("1. Start")
                            print("2. Stop")
                            choise=int(input("Enter your choise:"))

                            if choise==1:
                                start=datetime.now()
                                print("Start Time:",start.strftime("[%m/%d/%Y, %H:%M:%S]"))
                                print("Starting....\n")

                            elif choise==2:
                                end=datetime.now()
                                print("End Time:",end.strftime("[%m/%d/%Y, %H:%M:%S]"))
                                print("Ending....\n")

                                result=end-start
                                print("Diffrent Time:",result,"\n")
                                break

                            else:
                                print("Invelid Choise...")
    
                #Countdown timer
                    case 5:
                        import time
                        seconds=int(input("Enter time for Countdown timer in Secodes:"))

                        while seconds>=0:
                           print(f"Time left:{seconds} seconds",end="\r")
                           time.sleep(1)
                           seconds-=1

                        print("\nTime's up!")
                       


                    case 6:
                        break
                       
                    case _:
                        print("Invelid choise.....")

        case 7:
            break
```
