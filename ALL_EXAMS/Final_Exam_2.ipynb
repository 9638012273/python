```python
#  FINAL_EXAM_2  #
#================#

#  Link:-  #
#==========#
#https://docs.google.com/document/d/168_9tIsVSLM3qNNWqnATzRPEF7SAU8sazCW3Sg5PTBQ/edit?tab=t.0
#==========#

import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
from datetime import datetime
import os
from PIL import Image

class Fitness_tracker:
    def __init__(self):
        self.id=100
        self.workout_data=[]
        self.directory={}
        
    def Add_data(self,nm):
        self.name=nm
        self.DATE=[] 
        self.WORKOUT_TYPE=[] 
        self.DURATION=[] 
        self.BURNNED_CALARIOUS=[]   
            
        while True:
            self.date=input("Enter date (DD-MM-YYYY) :")
            self.workout_type=input("Enter Workout-type (cycling,running...):")
            self.duration=int(input("Enter Duration in minute:"))
            self.burned_calories=int(input("Enter burned calarious:"))
            
            self.DATE.append(self.date) 
            self.WORKOUT_TYPE.append(self.workout_type) 
            self.DURATION.append(self.duration) 
            self.BURNNED_CALARIOUS.append(self.burned_calories)

            ynchoise=input("Do you want to Add data (yes or no.):").lower()
            
            if ynchoise=="no":
                self.id+=1
                directory={
                    "id":self.id,
                    "name":self.name,
                    "date":self.DATE,
                    "workout_type":self.WORKOUT_TYPE,
                    "duration":self.DURATION,
                    "burnred_calarious":self.BURNNED_CALARIOUS,
                    }
                self.workout_data.append(directory)   
                print("\n\t\tProgramm Will continue...\n")
                break
            
            elif ynchoise!="yes":
                print("\n\t\tEnter only Yes or no...\n")
            

    def show_data(self):
        for i in self.workout_data:
            print(f"\nId: {i['id']}")
            print(f"Name: {i['name']}")
            print("-------------------->")
            print("Date        Workout          Duration       Calories")
            print("----------------------------------------------------")
            
            innerdata=zip(i['date'], i['workout_type'], i['duration'], i['burnred_calarious'])
            
            for dt,wk,du,ca in innerdata:
                print(f"{dt}  {wk}          {du} mins        {ca} calories")
            print("----------------------------------------------------\n\n") 


    def search_data(self):
        for i in self.workout_data:
            print("-----------------------")
            print(f"Id: {i['id']} --> Name: {i['name']} ")
            print("-----------------------")
        
        self.search_id=int(input("Enter Id to calculate burned calories:"))     
        total_calories=0
        
        for i in self.workout_data:
            if self.search_id==i['id']:
                print("\t.......................")
                print(f"\tId: {i['id']} --> Name: {i['name']} ")
                print("\t   * Burned Calories:-")
                innerdata=zip(i['date'], i['workout_type'], i['duration'], i['burnred_calarious'])
                for dt,wk,du,ca in innerdata:
                    print(f"\t   {ca}")
                    total_calories+=ca
        
        print("\t.......................")
        print(f"\tTotal Burned calories: {total_calories}")
        print("\t.......................")
        
        
    def Search_RegEx(self):
        if len(self.workout_data)==0:
            print("First enter data using Option 1 ....")
        
        else:
            search_date=input("Enter Date to Search Data: ")
            person={}
            total_calories=0 
            
            for record in self.workout_data:
                name=record["name"]
                for date,cal in zip(record["date"],record["burnred_calarious"]):
                    if date==search_date:
                        person[name]=person.get(name,0)+cal
                        total_calories+=cal

            if person:
                print(f"\nCalories burned on{search_date}:")
                for nm,total_cal in person.items():
                    print(f"{nm}:{total_cal} calories")

                print(f"\nTotal Calories burned on {search_date}:{total_calories} calories")
            else:
                print("Data not found!!!")

class Scraper:
    def  ScrapingData(self):
        print("\n\t\t Running...\n")
        print("link:-https://www.health.harvard.edu/category/staying-healthy")

        url="https://www.health.harvard.edu/category/staying-healthy"

        Title=[]
        Description=[]
        Prices=[]
        Image=[]

        request=requests.get(url)
        soup=BeautifulSoup(request.text,"html.parser")

        print("---------------------------------------------")
        print("Find All Title Name:-")
        print("---------------------------------------------")
        title=soup.find_all("p",class_="heading-subtle")
        for i in title:
            tt=i.text
            Title.append(tt)
        for i in Title:
            print(i)

        print("\n\n---------------------------------------------")
        print("Find All Title Description:-")
        print("---------------------------------------------")
        description=soup.find_all("div",class_="flex flex-col")
        for i in description:
            des=i.h3.text
            Description.append(des)
        for i in Description:
            print(i)

        print("\n\n---------------------------------------------")
        print("Find All Title Links:-")
        print("---------------------------------------------")
        price = soup.find_all("a", class_="article-card")
        for i in price:
            prc = i.get('href')
            Prices.append(prc)
        for prc in Prices:
            print(prc)


        print("\n\n---------------------------------------------")
        print("Find All Images:-")
        print("---------------------------------------------")
        img = soup.find_all("a", class_="article-card")
        for i in img:
            mg = i.img['src']
            Image.append(mg)
        for prc in Image:
            print(prc)
                

        makefile=input("Do You Want To save this data? (yes/no):")

        if(makefile=="yes"):
            filename=input("Enyter File name to save (e.g., data.csv):")
            dataframe=pd.DataFrame({"Title":Title,"description":Description,"link":Prices,"Images":Image,})
            dataframe.to_csv(filename)
            print(f"\n'{filename}' created successfully ...\n")

        elif(makefile=="no"):
            print("\nprogramm will be continue...\n")
        else:
            print("Enter input only yes or no...")
            
class Web_Automation:
    def  Form_fillup(self):
        f_name=input("Enter Your First name:")
        l_name=input("Enter Your Last name :")
        mobile=input("Enter Your Mobile no.:")
        email=input("Enter Your Email:")
        street=input("Enter street address:")
        city=input("Enter Your city:")
        state=input("Enter Your state:")
        gender=input("Enter Your gender:")
        dob=input("Enter your Birth Date (DD-MM-YYYY):")
        visit_date=input("Enter your Hoapital Visit Date (DD-MM-YYYY):")
        problem=input("Enter your problem:")

        date= datetime.strptime(dob, "%d-%m-%Y")
        dob_day = date.day
        dob_month = date.month
        dob_year = date.year

        datevisit= datetime.strptime(visit_date, "%d-%m-%Y")
        visit_day = date.day
        visit_month = date.month
        visit_year = date.year



        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service)
        driver.get(f"https://form.jotform.com/umaretiyamiit/personal-data-for-health-checkup")
        time.sleep(1)


        elem = driver.find_element(By.ID,"first_5")
        elem.send_keys(f_name)
        time.sleep(1)

        elem = driver.find_element(By.ID,"last_5")
        elem.send_keys(l_name)
        time.sleep(1)

        elem = driver.find_element(By.ID,"input_6_full")
        elem.send_keys(mobile)
        time.sleep(1)

        elem = driver.find_element(By.ID,"input_8")
        elem.send_keys(email)
        time.sleep(1)

        elem = driver.find_element(By.ID,"input_4_addr_line1")
        elem.send_keys(street)
        time.sleep(1)

        elem = driver.find_element(By.ID,"input_4_city")
        elem.send_keys(city)
        time.sleep(1)

        elem = driver.find_element(By.ID,"input_4_state")
        elem.send_keys(state)
        time.sleep(1)

        elem = driver.find_element(By.ID,"input_21")
        elem.send_keys(gender)
        time.sleep(1)

        elem = driver.find_element(By.ID,"input_22_day")
        elem.send_keys(dob_day)
        time.sleep(1)
        elem = driver.find_element(By.ID,"input_22_month")
        elem.send_keys(dob_month)
        time.sleep(1)
        elem = driver.find_element(By.ID,"input_22_year")
        elem.send_keys(dob_year)
        time.sleep(1)

        elem = driver.find_element(By.ID,"day_15")
        elem.send_keys(visit_day)
        time.sleep(1)
        elem = driver.find_element(By.ID,"month_15")
        elem.send_keys(visit_month)
        time.sleep(1)
        elem = driver.find_element(By.ID,"year_15")
        elem.send_keys(visit_year)
        time.sleep(1)

        elem = driver.find_element(By.ID,"input_24")
        elem.send_keys(problem)

        button=driver.find_element(By.ID,"input_25")
        elem= button.find_element(By.XPATH, "//button[@type='submit']")
        elem.send_keys(Keys.RETURN)

        Path=os.path.join(os.getcwd(),'healthdata_1.png')
        driver.save_screenshot(Path)

        driver.quit()

        print("\n\t\t==================================================================")
        print("\t\t\t \t HEALTH CARE DATA RECIVED ...")
        print("\t\t==================================================================")
        print(f"\t\t\t *** {f_name},{l_name} -->")
        print(f"\t\t\t Mobile no. : {mobile}")
        print(f"\t\t\t Email  Id. : {email}")
        print(f"\t\t\t Street Add.: {street}")
        print(f"\t\t\t City       : {city}")
        print(f"\t\t\t State      : {state}")
        print(f"\t\t\t Gender     : {gender}")
        print(f"\t\t\t DOB        : {dob}")
        print(f"\t\t\t Visit date : {visit_date}")
        print(f"\t\t\t Problem    : {problem}")
        print("\t\t==================================================================")
        print(f"\t\tForm Fill-up successfully....\n")
        

        current_directory = os.getcwd()
        image_path = os.path.join(current_directory, 'healthdata_1.png')
        img = Image.open(image_path)
        img.show()
        
#________________________________________________________________________________________________________________________________#
#                                    MAIN FUNCTION
#________________________________________________________________________________________________________________________________#
            

print("\n\t ______________________________________________________")
print("\t|  Welcome To Personalized Fitness And HEalth Tracker  |")
print("\t|______________________________________________________|\n")

obj=Fitness_tracker()
scraper_obj=Scraper()
automation_obj=Web_Automation()


while True:

    print("\n\t\t   * Main Menu:-")
    print("\t\t1. Workout & Fitness Tracker")
    print("\t\t2. Search Data using RegEx (Reg-Ex)")
    print("\t\t3. Scrape Health Tips From Website (Scrape-Data)")
    print("\t\t4. Do health Inquery Using Form Fill (Web-automation)")
    print("\t\t5. Exit\n")

    choice=int(input("Enter Your choice:"))
    match choice:
        case 1:
            while True:
                print("\n\t\t\t  * Chose An Option:-")
                print("\t\t\t1. Add Workout data")
                print("\t\t\t2. View Workouts")
                print("\t\t\t3. Get Total Calories burned")
                print("\t\t\t4. Back To Main Menu")

                health_choice=int(input("Enter Your Choise:"))
                match health_choice:
                    case 1:
                        name=input("Enter your name:")
                        obj.Add_data(name)
                    case 2:
                        obj.show_data()
                    case 3:
                        obj.search_data()
                    case 4:
                        print("\n\t\t\t Continue To Main Programm...\n")
                        break
                    case _:
                        print("\n\t\tInvelid choice...\n")
        
        case 2:
            obj.Search_RegEx()
            
        case 3:
            scraper_obj.ScrapingData()
            
        case 4:
            automation_obj.Form_fillup()
        
        case 5:
            print("\t\tExiting Programm....")
            break
        case _:
            print("\n\tInvelid choice...\n")
```
