```python
# Collection manipulator #
# ====================== #

print("Welcome the student data orgenizer!")

mainlist=[]  #list
dict={}      #dictionary

while True:     #loop continue while programm not break

    print("select an opstion:-\n")

    print("1. Add student")
    print("2. Display all students")
    print("3. Update student information")      #opstion for students 
    print("4. Delete student.")
    print("5. Display subject offered")
    print("6. Exit")

    choise=int(input("\n Enter your choise:"))      

    match choise:       

        #CASE 1 TAKE VALUE FROM USER

        case 1:
            num=int(input("how many students information add you:"))  

            print("Add student information:")

            for student in range(1,num+1):  

                print(f"STUDENT :{student}\n")
                id=int(input("Enter student id:"))
                name=input("Enter student name:")
                age=int(input("Enter student age:"))      #get student details from user
                grade=input("Enter student grade:")
                dob=input("Enter date of birth (yyyy-mm-dd):")
                sub=input("Subjects (coma sepreted):").split(',')

                dict[id]={   
                    "id":id,        #data convert in dictionary and id is key of dictionary 
                    "name":name,
                    "age":age,
                    "grade":grade,
                    "dob":dob,
                    "sub":sub,
                }
                mainlist.append(dict[id]) # dictionary append in main list(join)

            print("DATA INSERT SUCCESSFULLY....\n")

                
        #CASE 2 FOR DISPLAY ALL STUDENTS INFORMATION

        case 2:
            if len(mainlist)==0:    #main list is empty
                print("MAIN LIST IS EMPTY ... FIRST ENTER STUDENT DATA...")
            
            else:           #display all student details
                print(" === DISPLAY ALL STUDENTS === \n")
                for student in mainlist:
                    print(student)


        #CASE 3 FOR UPDATE STUDENT DETAILS      
     
        case 3:
            if len(mainlist)==0:
                print("MAIN LIST IS EMPTY ... FIRST ENTER STUDENT DATA...")
            
            else:

                user_id=int(input("Enter student id to update id details :"))
                
                for student in mainlist:
                    if student["id"]==user_id:

                        print("\n*** UPDATE DETAILS ***\n")

                        print("1. chenge ID")
                        print("2. chenge NAME")
                        print("3. chenge AGE")      #opstion for students 
                        print("4. chenge GRADE")
                        print("5. chenge DATE OF BIRTH")
                        print("6. chenge SUBJECT")
                        print("7. exit")

                        update_choise=int(input("\n Enter your choise:"))


                        if update_choise==1:
                            
                            for student in mainlist:          #match id in list dictionary
                                if student["id"]==user_id:     

                                    update_id=int(input("Update Student id:"))
                                    student["id"]=update_id  #update id via id

                                    print("SELECTED ID DETAILS UPDATED SUCCESSFULLY....\n")

                                else:
                                    print("ID IS NOT MATCH IN MAINLIST.....")



                        elif update_choise==2:

                            for student in mainlist:          #match id in list dictionary
                                if student["id"]==user_id:     

                                    update_name=input("Update Student Name:")
                                    student["name"]=update_name     #update name via id

                                    print("SELECTED ID NAME DETAILS UPDATED SUCCESSFULLY....\n")

                                else:
                                    print("ID IS NOT MATCH IN MAINLIST.....")

                        elif update_choise==3:

                            for student in mainlist:          #match id in list dictionary
                                if student["id"]==user_id:     

                                    update_age=int(input("Update Student Age:"))
                                    student["age"]=update_age     #update age via id

                                    print("SELECTED ID AGE DETAILS UPDATED SUCCESSFULLY....\n")

                                else:
                                    print("ID IS NOT MATCH IN MAINLIST.....")

                        elif update_choise==4:

                            for student in mainlist:          #match id in list dictionary
                                if student["id"]==user_id:     

                                    update_grade=input("Update Student Grade:")
                                    student["grade"]=update_grade     #update grade via id

                                    print("SELECTED ID GRADE DETAILS UPDATED SUCCESSFULLY....\n")

                                else:
                                    print("ID IS NOT MATCH IN MAINLIST.....")

                        elif update_choise==5:

                            for student in mainlist:          #match id in list dictionary
                                if student["id"]==user_id:     

                                    update_dob=input("Update Student Date of Birth:")
                                    student["dob"]=update_dob     #update dob via id

                                    print("SELECTED ID DATE OF BIRTH DETAILS UPDATED SUCCESSFULLY....\n")

                                else:
                                    print("ID IS NOT MATCH IN MAINLIST.....")

                        elif update_choise==6:

                            for student in mainlist:          #match id in list dictionary
                                if student["id"]==user_id:     

                                    update_sub=input("Update Subjects (coma sepreted):").split(',')
                                    student["sub"]=update_sub     #update subject via id

                                    print("SELECTED ID SUBJECT DETAILS UPDATED SUCCESSFULLY....\n")

                                else:
                                    print("ID IS NOT MATCH IN MAINLIST.....")
                                
                        elif update_choise==7:
                            print("Exiting system from update details\n")
                            continue
                            
                        else:
                            print("WRONG CHOISE...\n")

                    else:
                        print("ID IS NOT MATCH IN MAINLIST.....")

                


        #CASE 4 FOR DELETE STUDENT DETAILS

        case 4:
            if len(mainlist)==0:
                print("MAIN LIST IS EMPTY ... FIRST ENTER STUDENT DATA...")
            
            else:
                delete=int(input("Enter student id to delete details :"))

                if delete in dict:      #search id for delete idictionary in list
                    
                    for student in mainlist:
                        if student["id"]==delete:         #remove searched id dictionary from list
                            mainlist.remove(student)   
                            print("MATCHED ID DETAIL DELETED SUCCESSFULL....\n")

                else:
                     print("ID IS NOT MATCHED IN MAINLIST...\n")
                        

        #CASE 5 FOR OFFERED SUBJECT DATA

        case 5:
            if len(mainlist)==0:
                print("MAIN LIST IS EMPTY ... FIRST ENTER STUDENT DATA...")

            else:
                
                sub_list = []

                for student in mainlist:          
                    sub_list.extend(student["sub"]) 

                subset = list(set(sub_list)) 
                print(subset)


        #CASE 6 FOR EXIT

        case 6:
            print("THANK YOU For Coming...Exiting System...!!!")
            break       #break programm
        

        #CASE _ FOR INVELID CHOISE

        case _:
            print("INVELID CHOISE...!!!")
```
