```python
# OOP Wraper By Mit Patel #
#=========================#

mainlist=[]         #Empty list and dictionary to store data
dict={}

#base class
class Employee:
    def __init__(self):     #constructure to define variable
        
        self.empid=None
        self.name=None
        self.age=None
        self.salary=None

    def setdata(self):      #take dafault data from user
       
        self.name=input("Enter Name:")
        self.age=int(input("Enter Age:"))
        self.empid=input("Enter Employee id:")
        self.salary=float(input("Enter Salary"))

        dict[self.empid]={   
                "Role":"Employee",
                "empid":self.empid,        #data convert in dictionary and empid is key of dictionary 
                "name":self.name,
                "age":self.age,
                "salary":self.salary,
                }
        mainlist.append(dict[self.empid])   #data add in mainlist


#inherit class from Employee class
class Manager(Employee):
    
    def __init__(self):
        super().__init__()  #using super key take value and variable from top base class
        self.dept=None

    def setter_employee(self):  #take department for manager
        super().setdata()
        self.dept=input("Enter Department like (HR,Finance):")

        dict[self.empid].update({  
                "Role":"Manager",    
                "Department":self.dept,     #role data add and update selected value in dictionary 
                })


#inherit class from Employee and Manager class
class Developer(Manager,Employee):
   
    def __init__(self):
        super().__init__()  #using super key take value and variable from top base class
        self.leng=None

    def setter_developer(self):
        super().setdata()
        self.leng=input("Enter Programming Lenguage (Python/java):")

        dict[self.empid].update({  
                "Role":"Developer",
                "Lenguage":self.leng,   #role data add and update selected value in dictionary     
                })


d1=Developer()  #object 
while True:     #loop continue while True
    
    print("\n===  CHOSE OPERATIONS..  ===\n")   #operations

    print("1. Create A Developer")
    print("2. Create An Employee")
    print("3. Create A Manager")
    print("4. Show Details")
    print("5. Compare Salaries")
    print("6. Exit\n")

    choise=int(input("\nEnter your choise:"))   #take choise from user

    match choise:
        case 1:
            d1.setter_developer()   #call develapor setter method
            print("Developer data insert successfull..\n")

        case 2:
            d1.setter_employee()    #call employee setter method
            print("Employee data insert successfull..\n")

        case 3:
            d1.setdata()      #call manager setdata method
            print("Manager data insert successfull..\n")

        #   FOR DISPLAY ALL RECORDS
        case 4:
            if len(mainlist)==0:    #condition for empty list
                print("Mainlkist is Empty ..First Enter data ....")

            else:
                print("\n*** ALL MEMBERS ***\n")
                for data in mainlist:       #display all data in mainlist
                    print(data)

                print("\nCHOSE OPTION-")        #operations
                print("1. Developer")
                print("2. Employees")
                print("3. Manager")
                print("4. Exit\n")

                d_choise=int(input("Enter display choise:"))

                #using role display only selected roles data from main list
                if d_choise==1:     
                    print("*** DEVELOPERS ***")
                    for data in mainlist:
                        if data["Role"]=="Developer":
                            print(data)

                elif d_choise==2:
                    print("*** EMPLOYEES ***")
                    for data in mainlist:
                        if data["Role"]=="Employee":
                            print(data)

                elif d_choise==3:
                    print("*** MANAGERS ***")
                    for data in mainlist:
                        if data["Role"]=="Manager":
                            print(data)

                elif d_choise==4:   #continue programm
                    continue

                else:
                    print("Wrong choise..!!\n")
        #   FOR CALCULATE DIFFRENCE OF SALARY
        case 5:
            if len(mainlist)==0:
                print("Mainlkist is Empty ..First Enter data ....")

            else:
            
                print("\n *** COMPARE SALARIES ***\n")
                print("Choose Two Employees For Compare Salaries With ID")#take first and second id from user
                first_id = input("Enter First Employee's ID (e.g.,E123): ")
                second_id = input("Enter Second Employee's ID (e.g.,F456): ")

                print("\nComparing salaries ===>\n")
                
                if first_id in dict and second_id in dict:

                    #give value in variable to selected data records from dictionary
                    f_salary=dict[first_id]["salary"]   
                    f_name=dict[first_id]["name"]
                    f_role=dict[first_id]["Role"]

                    s_salary=dict[second_id]["salary"]
                    s_name=dict[second_id]["name"]
                    s_role=dict[second_id]["Role"]

                    class Complex:  #class for calculate salary

                        def __init__(self, f_salary,s_salary ): #give perameters value to variables
                            self.first=f_salary
                            self.second=s_salary

                        def __lt__(self):   #dunder method to calculate lessthen value
                            return self.first<self.second
                        

                    c1 = Complex(f_salary,s_salary) #object of class complex

                    print("===> CALCULATION <===")
                    print(f"{f_role} {f_name} (ID:{first_id})-----> {f_salary}")
                    print(f"{s_role} {s_name} (ID:{second_id})-----> {s_salary}")
                    print("===========================")

                    #catch return values
                    if c1.first>c1.second:  #calculate greater value 
                        print(f"{f_role} {f_name} (ID:{first_id}) has Greater salary then {s_role} {s_name} (ID:{second_id})")
                    
                    elif c1.first<c1.second:    #calculate less value 
                        print(f"{s_role} {s_name} (ID:{second_id}) has Greater salary then {f_role} {f_name} (ID:{first_id})")

                    else:   #else equal values
                        print(f"{f_role} {f_name} (ID:{first_id}) has Equal salary of {s_role} {s_name} (ID:{second_id})")

                else:
                    print("\n ID's Not Match In Dictionary ..Please check ID's")

        case 6:     #for exit
            print("Thank you for coming ... \n")
            break

        case _:
            print("WRONG CHOISE.....!!!\n")

```
