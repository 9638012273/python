```python
#que-9
#deposit and withdraw money

class Calculate():
    acno=None
    name=None
    money=None

    d_money=None
    w_money=None

    def get_data(m1):
        m1.acno=int(input("Enter your account no:"))
        m1.name=input("Enter your name:")
        m1.money=0

    def set_data(m1):
        print(f"A\c no. :{m1.acno} \nName :{m1.name} \nMoney :{m1.money}")

    def deposite(m1):
        m1.d_money=int(input("Enter deposite money:"))

        m1.money+=m1.d_money
        print(f"New balence :{m1.money}")
        print("Deposited Successfully")

    def withdraw(m1):
        m1.w_money=int(input("Enter withdrow money:"))

        if m1.w_money>m1.money:
            print(f"your amount is big .. noew balence in your account is {m1.money}")
        
        else:
            m1.money-=m1.w_money
            print(f"New balence :{m1.money}")
            print("Withdrow Successfully")
            

m1=Calculate()

m1.get_data()
m1.set_data()

while True:
    print("\n1. Deposite")
    print("2. Withdrow")
    print("3. Exit\n")
    choise=int(input("Enter your choise:"))
    
    if choise==1:
        m1.deposite()

    elif choise==2:
        m1.withdraw()

    elif choise==3:
        print("Thank you .... Exiting Programm...")
        break

    else:
        print("Wrong choise!!!")
```

```python
#que-11
#getter_setter & privatr_public

class Student():
    name=None
    sub1=None
    sub2=None
    sub3=None

    def getter(s1):
        s1.name=input("Enter Student Name:")
        s1.sub1=float(input("Enter Maths Marks:"))
        s1.sub2=float(input("Enter Science Marks:"))
        s1.sub3=float(input("Enter biology Marks:"))

    def setter(s1):
        print(f"\nStudent name:{s1.name}\n")
        print(f"Maths Marks:{s1.sub1}")
        print(f"Science Marks:{s1.sub2}")
        print(f"biology Marks:{s1.sub3}")

    def average(s1):
        s1.avg=(s1.sub1+s1.sub2+s1.sub3)/3

        print(f"\nAverage :{s1.avg:.2f}")

        if s1.avg>=90:
            print("Grade: A+")

        elif s1.avg>=80:
            print("Grade: B+")

        elif s1.avg>=70:
            print("Grade: C")

        elif s1.avg>=60:
            print("Grade: D")

        elif s1.avg>=34:
            print("Grade: E")

        else:
            print("Fail !...")




s1=Student()

s1.getter()
s1.setter()
s1.average()
```
