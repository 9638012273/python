```python
#QUE-8
#8. Python Program to Check ArmstrongÂ Number.

num=int(input("Enter the number for checking armstrong number:"))

temp=num
sum=0

while temp>0:
    digit=temp%10
    sum+=digit**3
    temp//=10
    
if num==sum:
   print(num,"is an armstrong number")
else:
   print(num,"is not an armstrong number") 
```

```python
#QUE-12
#12. Python Program that accepts a string and calculate the number of digits and letters.

string=input("Enter string:")

digit=0
latter=0

for x in string:
    if x.isalpha():
        latter+=1
    if x.isdigit():
        digit+=1

print("String is:",string,"\n")
print("Total latters in string =",latter)
print("Total digits in string =",digit)
```

```python
#QUE-13
#13. Python Program to check the validity of password input by users.
# Validation :
# a. At least 1 letter between [a-z] and 1 letter between [A-Z].
# b. At least 1 number between [0-9].
# c. At least 1 character from [$#@].
# d. Minimum length 6 characters.
# e. Maximum length 16 characters.

password=input("Enter the password:")

length=False
lower=False
upper=False
digit=False
special_symbol=False

if 6<= len(password) <=16:
    length=True
    
    for x in password:
        if x.islower():
            lower=True
        elif x.isupper():
            upper=True
        elif x.isdigit():
            digit=True
        elif x in "$#@":
            special_symbol=True
   

if (length and lower and upper and digit and special_symbol):
    print("PASSWORD IS VELID...")
else:
    print("PASSWORD IS NOT VELID...!!!")
```

```python
#QUE-14
#14. Python Program to check whether an alphabet is a vowel or consonant.

string=input("Enter the string:")

print("String is:",string)

for i in string.lower():
    if(i=="a" or i=="e" or i=="i" or i=="o" or i=="u"):
        print("Vowels:",i)
    else:
        print("Constants:",i)

```

```python
#QUE-15
#15. Python Program to convert month name to a number of days.
# Expected Output:
# List of months: January, February, March, April, May, June, July, August, September, October, November, December
# Input the name of Month: February
# No. of days: 28/29 days

month=input("Enter month name like(january/february...):")

match month.lower():
    case "january":
        print("Month : january")
        print("No. of days: 30 days")
    
    case "fabruary":
        print("Month : february")
        print("No. of days: 28/29 days")

    case "march":
        print("Month : march")
        print("No. of days: 31 days")

    case "april":
        print("Month : april")
        print("No. of days: 30 days")

    case "may":
        print("Month : may")
        print("No. of days: 31 days")

    case "june":
        print("Month : june")
        print("No. of days: 30 days")

    case "july":
        print("Month : july")
        print("No. of days: 31 days")
    
    case "auguest":
        print("Month : auguest")
        print("No. of days: 31 days")

    case "septmber":
        print("Month : septmber")
        print("No. of days: 30 days")

    case "october":
        print("Month : october")
        print("No. of days: 31 days")

    case "november":
        print("Month : november")
        print("No. of days: 30 days")

    case "december":
        print("Month : december")
        print("No. of days: 31 days")

    case _:
        print("WRONG CHOISE.....")


```
