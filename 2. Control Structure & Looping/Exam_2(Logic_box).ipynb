```python
# Logic box

print("Welcome to the pattern generator and number analyzer!")

while True:    # True: for loop continue while user not stop programm

    print("\nSelect an option:")
    print("1. Generate a pattern")
    print("2. Analyze a number")      #opstions
    print("3. Exit")

    choise=int(input("\nEnter your choise:")) #choise for pattern and number analyze
    sum=0

    match choise:
        case 1:

            print("\nChose a pattern type:")
            print("1. Right-angled tringle")
            print("2. Pyramid")                 #opstions
            print("3. Left-angled pattern")

            p_choise=int(input("\nEnter your choise:"))     #choise for pattern 

            match p_choise:
                case 1:

                    row=int(input("Enter the number of rows for the Right-angled pattern:"))
                    print("Pattern:")
                    
                    for i in range(1,row+1):
                        for j in range(1,i+1):
                            print("*", end = " ")
                        print()
                
                case 2:

                    row=int(input("Enter the number of rows for the Pyramid pattern:"))
                    print("Pattern:")
                    for i in range(1, row+1):
                        for j in range(row-i):
                            print(" ",end=" ")
                        for k in range(1,i+1):
                            print("*",end=" ")
                        for l in range(2,i+1):
                            print('*', end = " ")
                        print()
                    

                case 3:

                    row=int(input("Enter the number of rows for the left-angled pattern:"))
                    print("Pattern:")
                    for i in range(1, row+1):
                        for j in range(row-i):
                            print(" ",end=" ")
                        for k in range(1,i+1):
                            print("*",end=" ")
                        print()


                case _:
                    print("WRONG CHOISE ...!!!")
            
        case 2:
            s_range=int(input("Enter the start of range:"))
            e_range=int(input("Enter the end of range:"))

            for i in range(s_range,e_range+1):
                if (i%2==0):
                    print("Number",i,"is Even")
                else:
                    print("Number",i,"is Odd")

            for i in range(s_range,e_range+1):
                sum+=i
            print("The sum of all numbers from",s_range,"to",e_range,"is:",sum)

        case 3:
            print("Exiting the programm .Good bye!")
            break

        case _:
            print("invelid choise!!!!")
```
