```python
# OOP wraper #
#============#

mainlist=[]
dict={}

class Employee:
    def __init__(self):
        
        self.empid=None
        self.name=None
        self.age=None
        self.salary=None

    def setdata(self):
       
        self.name=input("Enter Name:")
        self.age=int(input("Enter Age:"))
        self.empid=int(input("Enter Employee id:"))
        self.salary=float(input("Enter Salary"))

        dict[self.empid]={   
                "Role":"Employee",
                "id":self.empid,        #data convert in dictionary and id is key of dictionary 
                "name":self.name,
                "age":self.age,
                "Salary":self.salary,
                }
        mainlist.append(dict[self.empid])

class Manager(Employee):
    
    def __init__(self):
        super().__init__()
        self.dept=None

    def setter_employee(self):
        super().setdata()
        self.dept=input("Enter Department like (HR,Finance):")

        dict[self.empid].update({  
                "Role":"Manager",    
                "Department":self.dept,     #data convert in dictionary and id is key of dictionary 
                })


class Developer(Manager,Employee):
   
    def __init__(self):
        super().__init__()
        self.leng=None

    def setter_developer(self):
        super().setdata()
        self.leng=input("Enter Programming Lenguage (Python/java):")

        dict[self.empid].update({  
                "Role":"Developer",
                "Lenguage":self.leng,
                })

   
d1=Developer()
while True:
    
    print("\n===  CHOSE OPERATIONS..  ===\n")

    print("1. Create A Developer")
    print("2. Create An Employee")
    print("3. Create A Manager")
    print("4. Show Details")
    print("5. Compare Salaries")
    print("6. Exit\n")

    choise=int(input("\nEnter your choise:"))

    match choise:
        case 1:
            d1.setter_developer() 
            print("Developer data insert successfull..\n")

        case 2:
            d1.setter_employee()
            print("Employee data insert successfull..\n")

        case 3:
            d1.setdata() 
            print("Manager data insert successfull..\n")

        case 4:
            print("\n*** ALL MEMBERS ***\n")
            for data in mainlist:
                print(data)

            print("\nCHOSE OPTION-")
            print("1. Developer")
            print("2. Employees")
            print("3. Manager\n")

            d_choise=int(input("Enter display choise:"))
            if d_choise==1:
                print("*** DEVELOPERS ***")
                for data in mainlist:
                    if data["Role"]=="Developer":
                        print(data)

            elif d_choise==2:
                print("*** EMPLOYEES ***")
                for data in mainlist:
                    if data["Role"]=="Employee":
                        print(data)

            elif d_choise==3:
                print("*** MANAGERS ***")
                for data in mainlist:
                    if data["Role"]=="Manager":
                        print(data)
            else:
                print("Wrong choise..!!\n")

        case 6:
            print("Thank you for coming ... \n")
            break

        case _:
            print("WRONG CHOISE.....!!!\n")

```
