```python
# RegEx Comander #
#=================#

#link - https://docs.google.com/document/d/1NIcETTr0KxGdJ6wFVzslhg-7N4IzURADhP3xgzD0irc/edit?tab=t.0



import re
import os

list_for_content=[]

Path=os.path.join(os.getcwd(),"data.txt")   
file=open(Path,"r")
total_data=file.read()   
file.close()

print("\n==========================================================")
print(" Welcome to the Pattern Extractor and Command-Line Utility ")
print("==========================================================")
while True:     #LOOP CONTINUE WHILE TRUE

    print("\n=======================================================")
    print("  ***   Chose an option:")
    print("--------------------------------------------------------")
    print("   1. Extract email addresses from tex")
    print("   2. Find all phone numbers matching a specific pattern ")      #OPSTIONS
    print("   3. Count occurrences of a specific word in text ")
    print("   4. Replace a pattern in the text with another string ")
    print("   5. Load text from a file ")
    print("   6. Save results to a new file ")
    print("   7. Exit")
    print("=======================================================\n")

    main_choise=int(input("Enter your choise:"))

    match main_choise:    
    
    #CASE 1 FOR EXTRACT ADDRESS
        case 1:
            # print("-----------------------------------")
            # print("Original Data: \n",total_data,"\n")
            # print("-----------------------------------")

            print("__Extracting email address...\n")   

            find_email=re.findall(r"[\w.]+@\w*.[A-Za-z]{2,}", total_data)     #find email using pattern

            if find_email:
                print("Found email address:")
                nm=1
                for email in find_email:
                    print(f"  {nm}. {email}")       #print all find emails with number counting
                    nm+=1
                
            else:
                print("Email address not found!!!")


    #CASE 2 FOR FIND DATA BY FORMETS
        case 2:
            print("-----------------------------------")
            print("Original Data: ",total_data,"\n")
            print("-----------------------------------\n")
            print("-Chose Formet:")
            print("1. 123-456-7890")          
            print("2. (555) 123-4567")
            print("3. 12345 67890")

            ch=int(input("Enter your choise:"))

            if ch==1:                                           #give pattern according choice
                pattern=r"\d{3}-\d{3}-\d{4}"        #for  123-456-7890
                found=True

            elif ch==2:
                pattern=r"[(]\d{3}[)].\d{3}-\d{4}"      #for  (555) 123-4567
                found=True

            elif ch==3:
                pattern=r"\d{5}.\d{5}"              #for  12345 67890
                found=True

            else:
                pattern=r"`"
                found=False
                print("Invalid choise..")
                

            if found==True:                         
                lst=[]
                find_phone=re.findall(pattern, total_data)    #find number from data

                for number in find_phone:
                    lst.append(number)      #finded number append to list

                if len(lst)==0:
                    print("\nPhone number not found!!!\n")

                else:
                    print("- Extracting phone numbers...\n")    #print all numbers from list
                    print("Found phone number:")
                    num=1
                    for number in lst:
                        print(f"  {num}. {number}")
                        num+=1

        
    #CASE 3 FOR FIND word in data
        case 3:
            print("-----------------------------------")
            print("Original Data: ",total_data,"\n")
            print("-----------------------------------\n")
            word=input("Enter the word for count:")

            if word in total_data:
                find_word=re.finditer(word, total_data)               #find word from data                

                print(f"- Counting occurance of {word}...")
                cnt=0
                for words in find_word:                          #print all count number for word find
                    cnt+=1

                print(f"\nThe Word '{word}' appers {cnt} times.")

            else:
                print("Word not found in data!!!")


    #CASE 4 REPLACE OLD PATTERN TO NEW PATTERN
        case 4:
            print("-----------------------------------")
            print("Original Data:\n",total_data,"\n")   
            print("-----------------------------------\n")  
            old_pattern=input("Enter the pattern to replace:")  #take replace pattern

            print("__Replacing pattern...\n")  
            if old_pattern in total_data:
                new_pattern=input("Enter the replacement string:")      #take new pattern
                new_data=re.sub(old_pattern,new_pattern,total_data)           #replace old pattern to new pattern using sub
                print(f"Updated Data:\n {new_data}")
                
            else:
                print(f"Pattern '{old_pattern}' not found in data!!!")


    #CASE 5 FOR LOAD DATA FROM FILE
        case 5:
            try:
                filename=input("Enter Filename To Read File:")

                file=open(os.path.join(os.getcwd(),filename),"r")       #file open in current path
                list_for_content.append(file.read())
                for content in list_for_content:        #all content append in list
                    print(content)
                file.close()
                print(f"\nFile content loded success fully with path:",os.getcwd())
                
            except FileNotFoundError:
                print("error: File not Found!!!")

    #CASE 6 FOR PASTE LODED DATA
        case 6:
            if len(list_for_content)==0:                    #if content not lodeed give error
                print("First load data From Another file...\noption 5: for load text....\n")

            else:
                try:
                    filename=input("Enter Filename To Create and past content in File:")

                    file=open(os.path.join(os.getcwd(),filename),"w")       #file open in current path
                    for content in list_for_content:
                        print(file.write(content))              #write all lodead content into file
                    file.close()
                    print(f"File content Written succcessfully... with path:",os.getcwd())

                except FileNotFoundError:
                    print("error: File not Found!!!")


    #CASE 7 FOR EXIT 
        case 7:
            print("Exiting the programm. Good bye...\n")
            break

    #INVELID CHOISE
        case _:
            print("Invalid choice...\n")
```
