```python
# OOP wrapper #
#============#

mainlist = []

class Employee:
    def __init__(self):
        self.empid = None
        self.name = None
        self.age = None
        self.salary = None
        self.id = 1

    def setdata(self):
        self.name = input("Enter Name: ")
        self.age = int(input("Enter Age: "))
        self.empid = int(input("Enter Employee id: "))
        self.salary = float(input("Enter Salary: "))

        employee_data = {
            "empid": self.empid,
            "name": self.name,
            "age": self.age,
            "Salary": self.salary,
        }
        mainlist.append((self.id, employee_data))

class Manager(Employee):
    def __init__(self):
        super().__init__()
        self.dept = None
        self.id = 2

    def setdata(self):
        super().setdata()
        self.dept = input("Enter Department like (HR, Finance): ")

        manager_data = {
            "empid": self.empid,
            "name": self.name,
            "age": self.age,
            "Salary": self.salary,
            "Department": self.dept,
        }
        mainlist.append((self.id, manager_data))

class Developer(Employee):
    def __init__(self):
        super().__init__()
        self.leng = None
        self.id = 3

    def setdata(self):
        super().setdata()
        self.leng = input("Enter Programming Language (Python/Java): ")

        developer_data = {
            "empid": self.empid,
            "name": self.name,
            "age": self.age,
            "Salary": self.salary,
            "Language": self.leng,
        }
        mainlist.append((self.id, developer_data))

def display_details():
    print("CHOOSE OPTION-")
    print("1. Developer")
    print("2. Employees")
    print("3. Manager\n")

    d_choice = int(input("Enter display choice: "))
    
    for (id, data) in mainlist:
        if id == d_choice:
            print(data)

developer = Developer()
employee = Employee()
manager = Manager()

while True:
    print("\n===  CHOOSE OPERATIONS..  ===\n")
    print("1. Create a developer")
    print("2. Create an employee")
    print("3. Create a manager")
    print("4. Show details")
    print("5. Exit")

    choice = int(input("\nEnter your choice: "))

    match choice:
        case 1:
            developer.setdata()
        case 2:
            employee.setdata()
        case 3:
            manager.setdata()
        case 4:
            display_details()
        case 5:
            print("Thank you for coming...")
            break
        case _:
            print("Invalid choice, please try again.")
```
